### Problem 1

##### Find the total number of crimes recorded in the CRIME table

```python
# Rows in Crime table
%sql select count(ID) as "ID" from CHICAGO_CRIME_DATA
```

### Problem 2

##### Retrieve first 10 rows from the CRIME table

```python
%sql SELECT * FROM CHICAGO_CRIME_DATA\
LIMIT 10
```

### Problem 3

##### How many crimes involve an arrest?

```python
%sql SELECT COUNT(*) FROM CHICAGO_CRIME_DATA\
WHERE ARREST='TRUE'
```

### Problem 4

##### Which unique types of crimes have been recorded at GAS STATION locations?

```python
%sql select DISTINCT(primary_type), location_description from CHICAGO_CRIME_DATA where location_description = 'GAS STATION'
```

### Problem 5

##### In the CENUS_DATA table list all Community Areas whose names start with the letter ‘B’.

```python
%sql select COMMUNITY_AREA_NAME from CENSUS_DATA where COMMUNITY_AREA_NAME LIKE 'B%'
```

### Problem 6

##### Which schools in Community Areas 10 to 15 are healthy school certified?

```python

%sql select NAME_OF_SCHOOL from SCHOOL where HEALTHY_SCHOOL_CERTIFIED=True and COMMUNITY_AREA_NUMBER between 10 and 15
```

### Problem 7

##### What is the average school Safety Score? 

```python

%sql select avg(SAFETY_SCORE) as "avg score" from SCHOOL
```

### Problem 8

##### List the top 5 Community Areas by average College Enrollment [number of students] 

```python

%sql select COMMUNITY_AREA_NAME , avg(COLLEGE_ENROLLMENT) as AVG_ENROLLMENT from SCHOOL group by COMMUNITY_AREA_NAME order by AVG_ENROLLMENT desc LIMIT 5
```

### Problem 9

##### Use a sub-query to determine which Community Area has the least value for school Safety Score? 

```python
%sql select COMMUNITY_AREA_NAME, SAFETY_SCORE from SCHOOL where SAFETY_SCORE=(select min(SAFETY_SCORE) from SCHOOL)
```

### Problem 10

##### [Without using an explicit JOIN operator] Find the Per Capita Income of the Community Area which has a school Safety Score of 1.

```python
%sql select CE.COMMUNITY_AREA_NAME, CE.PER_CAPITA_INCOME from SCHOOL as CPS, CENSUS_DATA as CE where CE.COMMUNITY_AREA_NUMBER = CPS.COMMUNITY_AREA_NUMBER and CPS.SAFETY_SCORE = 1
```
